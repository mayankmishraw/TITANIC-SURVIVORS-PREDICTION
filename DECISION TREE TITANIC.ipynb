```python
df=pd.read_csv("titanic.csv")
```

```python
df
```

```python
df.drop(["PassengerId","Name","SibSp","Parch","Ticket","Cabin","Embarked"],axis=1,inplace=True)
```

```python
df
```

```python
fill=df["Age"].mean()
```

```python
df["Age"].fillna(fill,inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df=pd.get_dummies(df,drop_first=True)  #we can also use dummy method for separate columns rather then whole dataset
```

```python
df
```

```python
df.rename(columns={"Sex_male":"male"},inplace=True)
```

```python
df
```

```python
y=df.Survived
x=df.drop("Survived",axis=1)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=10)
```

```python
xtrain
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
ds=DecisionTreeClassifier()
```

```python
ds.fit(xtrain,ytrain)
```

```python
pred=ds.predict(xtest)
```

```python
pred
```

```python
ytest
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(pred,ytest)
```

```python
ds.score(xtest,ytest)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(pred,ytest)
```
